```python
def shorten_labels(df):
    def path(x):
        return str(x)[27:]
    
    df['labels'] = df['labels'].apply(path)

    return df

#""" 
# USED FOR 100k
def add_pathname(df, num):
    def path(x):
        return '../cartoon/cartoonset100k/'+str(num)+'/cs' + str(x) + '.png'
    
    df.insert(loc=1, column='path', value=df['labels'].apply(path))

    return df
#"""
"""
# USED FOR 10k
def add_pathname(df, num):
    def path(x):
        return '../cartoon/cartoonset10k/cs' + str(x) + '.png'
    
    df.insert(loc=1, column='path', value=df['labels'].apply(path))

    return df
"""

def sumarize_face_color(x):
    if x in [4, 5]:
        return int(4)
    if x in [6, 7, 8]:
        return int(5)
    if x in [9, 10]:
        return int(6)
    else:
        return int(x)

   
def sumarize_hair(x):
    if x in [0, 1, 2, 108, 110]:
        return int(0)
    if x in [18, 19, 20, 21, 22, 109, 110]:
        return int(1)
   
    if x in [17, 32, 44, 49, 61, 76, 83]:
        return int(2)
    if x in [11, 12, 23, 31, 39]:
        return int(3)
    
    if x in [10, 24, 30, 36, 37, 38, 88, 98, 104]:
        return int(4)
    if x in [43, 45, 46, 50, 52, 54, 62, 77]:
        return int(5)
    if x in [89, 92, 99, 105, 107]:
        return int(6)
    if x in [51, 53, 66, 72, 73, 74]:
        return int(7)
    if x in [60, 65, 67, 71, 75, 80, 81]:
        return int(8)
    
    if x in [86, 90, 91, 94, 97, 101, 103]:
        return int(9)
    if x in [6, 7, 13, 16]:
        return int(10)
    if x in [9, 26, 34, 106]:
        return int(11)
    if x in [40, 42, 47, 57, 63, 69, 79, 82]:
        return int(12)
    
    if x in [25, 33, 78, 85, 87, 93, 100]:
        return int(13)
    
    if x in [3, 4, 28, 29, 41, 96]:
        return int(14)
    if x in [27, 35, 48, 56, 58, 59, 64, 70, 87, 95, 102]:
        return int(15)
    if x in [55, 68, 84]:
        return int(16)
    
    
    if x in [8, 14]:
        return int(17)
    if x in [5, 15]:
        return int(18)
    
    else:
        return int(x)


def sumarize_facial_hair(x):
    if x in [1, 14]:
        return int(0)
    if x in [4, 5, 6, 7, 12, 13]:
        return int(1)
    if x in [0, 8, 9, 10]:
        return int(2)
    if x in [2, 3, 11]:
        return int(3)
    else:
        return int(x)


def sumarize_hair_color(x):
    if x == 8:
        return 9-int(0)
    if x == 9:
        return 9-int(1)
    if x == 0:
        return 9-int(2)
    if x == 1:
        return 9-int(3)
    if x == 4:
        return 9-int(4)
    if x == 2:
        return 9-int(5)
    if x == 3:
        return 9-int(6)
    if x == 5:
        return 9-int(7)
    if x == 6:
        return 9-int(8)
    if x == 7:
        return 9-int(9)
    else:
        return 9-int(x)
```

```python
def apply(num):
    df = pd.read_feather('../cartoon/meta100k/org_meta/cartoon100k_'+ str(num) +'meta.feather')
   
    df = shorten_labels(df)
    df = add_pathname(df, num)

    #attr_dict['nhair'] = (19, 'Hair')
    df.insert(loc=0, column='nhair', value=df['hair'].apply(sumarize_hair))

    #attr_dict['nface_color'] = (7, 'Face color') #attr_dict.update({'nface_color': (7, 'Face color')}) 
    df.insert(loc=1, column='nface_color', value=df['face_color'].apply(sumarize_face_color))

    #attr_dict['nhair_color'] = (10, 'Hair color') 
    df.insert(loc=2, column='nhair_color', value=df['hair_color'].apply(sumarize_hair_color))

    #attr_dict['nfacial_hair'] = (4, 'Facial Hair')
    df.insert(loc=3, column='nfacial_hair', value=df['facial_hair'].apply(sumarize_facial_hair))
    
    df.to_feather('../cartoon/meta100k/cartoon100k_'+ str(num) +'_meta.feather')
```
